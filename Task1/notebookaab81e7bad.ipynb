```python
import pandas as pd

file_path = "/kaggle/input/datasets/chanadanalucky/dataset/API_SP.POP.TOTL_DS2_en_csv_v2_61.csv"

df = pd.read_csv(file_path, skiprows=4)

df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

latest_year = "2022"

population_data = df[latest_year].dropna()

plt.figure()
plt.hist(population_data, bins=20)

plt.title("Distribution of Population Across Countries (2022)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
```

```python
latest_year = "2022"

top10 = df[["Country Name", latest_year]].dropna().sort_values(by=latest_year, ascending=False).head(10)

plt.figure()
plt.bar(top10["Country Name"], top10[latest_year])

plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()
```
